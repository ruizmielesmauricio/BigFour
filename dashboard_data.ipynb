```python
import pandas as pd

df = pd.read_csv("big4_tracks_with_bpm_clean.csv")

df.head()
```

```python
df = df[(df["album_name"] != "Crazy Sheep") & 
        (df["album_name"] != "Dead Commander")]
```

```python
top_tracks = df.sort_values(
    "lastfm_playcount", ascending=False
).groupby("artist").head(5)

top_tracks
```

```python
album_stats = df.groupby(
    ["artist", "album_name", "first_release_date"]
).agg(
    total_tracks=("track_name", "count"),
    total_playcount=("lastfm_playcount", "sum"),
    total_listeners=("lastfm_listeners", "sum"),
    avg_bpm=("bpm", "mean"),
    avg_track_length=("track_length_ms", "mean")
).reset_index()

album_stats
```

```python
top_albums = album_stats.sort_values(
    "total_playcount", ascending=False
).groupby("artist").head(1)

top_albums
```

```python
def bpm_category(bpm):
    if pd.isna(bpm):
        return "Unknown"
    elif bpm < 100:
        return "Slow"
    elif bpm < 140:
        return "Medium"
    else:
        return "Fast"

df["bpm_category"] = df["bpm"].apply(bpm_category)
```

```python
album_bpm_profile = df.groupby(
    ["artist", "album_name", "bpm_category"]
).size().reset_index(name="count")
```

```python
album_stats.to_csv("big4_album_stats.csv", index=False)
top_tracks.to_csv("big4_top_tracks.csv", index=False)
top_albums.to_csv("big4_top_albums.csv", index=False)
```

```python
## Albums with no BPM
album_bpm_check = df.groupby(["artist", "album_name"]).agg(
    total_tracks=("track_name", "count"),
    tracks_with_bpm=("bpm", lambda x: x.notna().sum()),
    avg_bpm=("bpm", "mean")
).reset_index()

album_bpm_check[album_bpm_check["avg_bpm"].isna()]
```

```python
## Songs with ni BPM
album_bpm_check["bpm_coverage"] = (
    album_bpm_check["tracks_with_bpm"] / album_bpm_check["total_tracks"]
)

album_bpm_check.sort_values("bpm_coverage").head(20)
```

```python
## Missing tracks BPM
album_bpm_check = df.groupby(
    ["artist", "album_name", "first_release_date"]
).agg(
    total_tracks=("track_name", "count"),
    tracks_with_bpm=("bpm", lambda x: x.notna().sum()),
    avg_bpm=("bpm", "mean")
).reset_index()

album_bpm_check["bpm_coverage"] = (
    album_bpm_check["tracks_with_bpm"] / album_bpm_check["total_tracks"]
)
```

```python
## Confidence
def coverage_label(x):
    if x >= 0.8:
        return "High"
    elif x >= 0.5:
        return "Medium"
    else:
        return "Low"

album_bpm_check["bpm_confidence"] = album_bpm_check["bpm_coverage"].apply(coverage_label)
```

```python
album_bpm_check.to_csv("big4_album_bpm_coverage.csv", index=False)
```
